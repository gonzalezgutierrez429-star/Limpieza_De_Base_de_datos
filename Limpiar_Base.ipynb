Creamos una funcion para limpiar datos

```python
def clean_value(value):
    if pd.isna(value):
        return value
    value_str = str(value).strip()
    if value_str in ['Auto%#', '', ' ', 'nan', 'NaN', 'NULL', 'None']:
        return np.nan
    return value_str

for col in df.columns:
    df[col] = df[col].apply(clean_value)
```

Aplicamos correcciones

```python
correcciones = {
    'Deductible': {'Auto%#': np.nan},
    'VehiclePrice': {'Auto%#': np.nan},
    'DriverRating': {'Auto%#': np.nan},
    'RepNumber': {'Auto%#': np.nan},
    'AccidentArea': {'Auto%#': np.nan},
    'Make': {'Auto%#': np.nan},
    'AgentType': {'Auto%#': np.nan},
    'BasePolicy': {'Auto%#': np.nan},
    'PoliceReportFiled': {'Auto%#': np.nan},
    'WitnessPresent': {'Auto%#': np.nan}
}

for col, corrections in correcciones.items():
    if col in df.columns:
        df[col] = df[col].replace(corrections)
```

Reviso si hay duplicados

```python
duplicados = df.duplicated().sum()
print('Duplicados encontrados:', duplicados)
df_sin_dup = df.drop_duplicates()
print('Después de eliminar duplicados:', len(df_sin_dup))
```

Traducimos los datos al español

```python
traducciones = {
    'Month': {'Jan': 'Enero', 'Feb': 'Febrero', 'Mar': 'Marzo', 'Apr': 'Abril', 'May': 'Mayo', 'Jun': 'Junio', 
              'Jul': 'Julio', 'Aug': 'Agosto', 'Sep': 'Septiembre', 'Oct': 'Octubre', 'Nov': 'Noviembre', 'Dec': 'Diciembre'},
    'DayOfWeek': {'Monday': 'Lunes', 'Tuesday': 'Martes', 'Wednesday': 'Miércoles', 'Thursday': 'Jueves', 
                  'Friday': 'Viernes', 'Saturday': 'Sábado', 'Sunday': 'Domingo'},
    'Sex': {'Male': 'Masculino', 'Female': 'Femenino'},
    'MaritalStatus': {'Single': 'Soltero', 'Married': 'Casado', 'Widow': 'Viudo'},
    'AccidentArea': {'Urban': 'Urbano', 'Rural': 'Rural'},
    'PoliceReportFiled': {'Yes': 'Sí', 'No': 'No'},
    'WitnessPresent': {'Yes': 'Sí', 'No': 'No'}
}

for col, trad in traducciones.items():
    if col in df_sin_dup.columns:
        df_sin_dup[col] = df_sin_dup[col].replace(trad)


print('Valores traducidos')
```

```python
nombres_español = {
    'Month': 'Mes', 'WeekOfMonth': 'SemanaDelMes', 'DayOfWeek': 'DiaDeLaSemana',
    'Make': 'MarcaVehiculo', 'AccidentArea': 'AreaAccidente', 'DayOfWeekClaimed': 'DiaReclamo',
    'MonthClaimed': 'MesReclamo', 'WeekOfMonthClaimed': 'SemanaReclamo', 'Sex': 'Genero',
    'MaritalStatus': 'EstadoCivil', 'Age': 'Edad', 'Fault': 'Culpa', 'PolicyType': 'TipoPoliza',
    'VehicleCategory': 'CategoriaVehiculo', 'VehiclePrice': 'PrecioVehiculo', 
    'FraudFound_P': 'FraudeDetectado', 'PolicyNumber': 'NumeroPoliza', 'RepNumber': 'NumeroRepresentante',
    'Deductible': 'Deducible', 'DriverRating': 'CalificacionConductor', 
    'Days_Policy_Accident': 'DiasPolizaAccidente', 'Days_Policy_Claim': 'DiasPolizaReclamo',
    'PastNumberOfClaims': 'NumeroReclamosAnteriores', 'AgeOfVehicle': 'EdadVehiculo',
    'AgeOfPolicyHolder': 'EdadAsegurado', 'PoliceReportFiled': 'ReportePolicial',
    'WitnessPresent': 'TestigoPresente', 'AgentType': 'TipoAgente',
    'NumberOfSuppliments': 'NumeroSuplementos', 'AddressChange_Claim': 'CambioDireccionReclamo',
    'NumberOfCars': 'NumeroVehiculos', 'Year': 'Año', 'BasePolicy': 'PolizaBase'
}

df_sin_dup = df_sin_dup.rename(columns=nombres_español)
print('Nombres de columnas cambiados a español')
```

Definimos una funcion para detectar y corregir valores (outliers)

```python
def corregir_outliers(columna):
    if columna.dtype in ['int64', 'float64']:
        Q1 = columna.quantile(0.25)
        Q3 = columna.quantile(0.75)
        IQR = Q3 - Q1
        li = Q1 - 1.5 * IQR
        ls = Q3 + 1.5 * IQR
        columna = np.where(columna < li, li, columna)
        columna = np.where(columna > ls, ls, columna)
    return columna

columnas_outliers = ['Edad', 'Deducible', 'CalificacionConductor']
for col in columnas_outliers:
    if col in df_sin_dup.columns:
        df_sin_dup[col] = corregir_outliers(df_sin_dup[col])

print('Valores atípicos corregidos')
```

creo una copia del DataFrame limpio y revisamos su tamaño

```python
df_limpio = df_sin_dup.copy()

print('Resumen:')
print('Dimensiones:', df_limpio.shape)
print('Registros:', len(df_limpio))
print('Columnas:', len(df_limpio.columns))
print('Duplicados:', df_limpio.duplicated().sum())
print('Valores faltantes totales:', df_limpio.isnull().sum().sum())
print('Tipos de datos:')
print(df_limpio.dtypes.value_counts())

print('\nPrimeras filas del dataset limpio:')
print(df_limpio.head(2))
```

Eliminamos filas sin datos

```python
#  Quitar filas con datos críticos faltantes
print('Quitando filas con datos importantes faltantes...')

filas_inicial = len(df_limpio)

# Quitar filas sin número de póliza
sin_poliza = df_limpio['NumeroPoliza'].isna().sum()
df_limpio = df_limpio[df_limpio['NumeroPoliza'].notna()]
print(f'Quitadas {sin_poliza} filas sin número de póliza')

# Quitar filas sin marca de vehículo
sin_marca = df_limpio['MarcaVehiculo'].isna().sum()
df_limpio = df_limpio[df_limpio['MarcaVehiculo'].notna()]
print(f'Quitadas {sin_marca} filas sin marca de vehículo')

# Quitar filas sin edad
sin_edad = df_limpio['Edad'].isna().sum()
df_limpio = df_limpio[df_limpio['Edad'].notna()]
print(f'Quitadas {sin_edad} filas sin edad')

filas_final = len(df_limpio)
print(f'Total filas quitadas: {filas_inicial - filas_final}')
```

```python
print('Buscando valores Auto%#...')

filas_con_autop = 0
for col in df_limpio.columns:
    if df_limpio[col].dtype == 'object':
        mask = df_limpio[col].astype(str).str.contains('Auto%#', na=False)
        if mask.any():
            count = mask.sum()
            filas_con_autop += count
            df_limpio = df_limpio[~mask]
            print(f'Quitadas {count} filas con Auto%# en {col}')

print(f'Total filas con Auto%# quitadas: {filas_con_autop}')
```

Quitamos filas con errores

```python
# Quitar filas con otros valores malos
print('Quitando filas con otros problemas...')

# Quitar filas sin mes
sin_mes = df_limpio['Mes'].isna().sum()
df_limpio = df_limpio[df_limpio['Mes'].notna()]
print(f'Quitadas {sin_mes} filas sin mes')

# Quitar filas sin área de accidente
sin_area = df_limpio['AreaAccidente'].isna().sum()
df_limpio = df_limpio[df_limpio['AreaAccidente'].notna()]
print(f'Quitadas {sin_area} filas sin área de accidente')

print(f'Filas restantes: {len(df_limpio)}')
```

Revisamos si aun quedan NaN

```python
print('Revisando qué falta limpiar...')

nan_restantes = df_limpio.isnull().sum()
columnas_con_nan = nan_restantes[nan_restantes > 0]

if len(columnas_con_nan) > 0:
    print('Columnas que todavía tienen NaN:')
    for col, count in columnas_con_nan.items():
        print(f'  {col}: {count} valores faltantes')
        
    # Para columnas con pocos NaN, llenar con valores comunes
    for col, count in columnas_con_nan.items():
        if count < 10:  # Si hay menos de 10 NaN
            if df_limpio[col].dtype in ['int64', 'float64']:
                valor = df_limpio[col].median()
                df_limpio[col].fillna(valor, inplace=True)
            else:
                valor = df_limpio[col].mode()[0] if not df_limpio[col].mode().empty else 'No especificado'
                df_limpio[col].fillna(valor, inplace=True)
            print(f'Llenados {count} NaN en {col} con: {valor}')
else:
    print('Ya no hay NaN en el dataset')
```
